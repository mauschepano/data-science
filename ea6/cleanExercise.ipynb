```python
import pandas as pd
import numpy as np
url="https://raw.githubusercontent.com/edlich/eternalrepo/master/DS-WAHLFACH/dsm-beuth-edl-demodata-dirty.csv"
df=pd.read_csv(url)
df
```

```python
df2 = df
print('Remove rows where every single data is missing.')
df3 = df2.dropna(axis=0, how='all')
df3
```

```python
print('Delete duplicates.')
df4 = df3.copy()
colsToCheck = ['full_name', 'first_name', 'last_name', 'email', 'gender', 'age']
df4.drop_duplicates(colsToCheck,keep = 'first', inplace = True)
df4
```
